# Explanation


1- For m, we will use data.shape[0] to get the number of records in the given data
2- For summation(x) we will use np.sum(x) to summate all the records of input data
3- For summation(y) we will use np.sum(y) to summate all the records of output data
4- For summation(x**2) we will use np.sum(np.dot(x,x)) to first square the x column and then summate it 
5- For summation(x,y) we will use np.sum(np.dot(x,y)) to first multiply the input x and output y columns respectively and then    summate the results 
6- Using np.array([]) to make the matrices according to the derived format

```python
#Your code Here
#For loop is not recommended in these cases, 
#use numpy functions to calculate the values of above variables in one line of code.
#Such as x.shape for number of rows, sum() for sum of all elements in x
#np.dot() or multiply suntion for square of x 
m=traindata.shape[0]
sum_x=np.sum(x_train)
#print(sum_x)
sum_xsquare=np.sum(np.dot(x_train,x_train))
#print(sum_xsquare)
sum_y=np.sum(y_train)
#print(sum_y)
sum_xy=np.sum(np.dot(x_train,y_train))
#print(sum_xy)
A = np.array([[m,sum_x],[sum_x,sum_xsquare]])
print("Matrix A : \n\n",A)
B = np.array([[sum_y],[sum_xy]])
print("\nMatrix B : \n\n",B)
```

3- Since the matrix is there, we can extract the values of theta_0 and theta_1 from their indexes.
